## Spliting Data and Scale/Tranform

```python
X = df.drop('benign_0__mal_1', axis = 1).values # to np array
y = df['benign_0__mal_1'].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=101)
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
scaler = MinMaxScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_test = scaler.fit_transform(X_test)
```

## Model

```python
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
```

```python
X_train.shape
```

```python
model = Sequential()
model.add(Dense(30, activation='relu'))
model.add(Dense(30, activation='relu'))

# binary classification
model.add(Dense(1, activation='sigmoid'))
model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
model.fit(x= X_train, y=y_train, epochs = 600,validation_data=(X_test, y_test))
```

```python
losses = pd.DataFrame(model.history.history)
```

```python
losses
```

```python
losses.plot()
```

# Early Stopping

```python
# dÃ¹ng early stoping Ä‘á»ƒ dá»«ng láº¡i khi cÃ³ dáº¥u hiá»‡u overfitting
model = Sequential()
model.add(Dense(30, activation='relu'))
model.add(Dense(30, activation='relu'))

# binary classification
model.add(Dense(1, activation='sigmoid'))
model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
from tensorflow.keras.callbacks import EarlyStopping
```

```python
help(EarlyStopping)
```

```python
early_stop = EarlyStopping(monitor= 'val_loss', mode = 'min', verbose = 1, patience = 25) # it mean it will stop when accurancy is not decrease
##One of `{"auto", "min", "max"}`. In `min` mode,
 #       training will stop when the quantity
 #       monitored has stopped decreasing; in `"max"`
   #     mode it will stop when the quantity
  #      monitored has stopped increasing; in `"auto"`
 #      mode, the direction is automatically inferred
  #      from the name of the monitored quantity.
```

```python
model.fit(x= X_train, y=y_train, epochs = 600,validation_data=(X_test, y_test),
         callbacks = [early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

# Dropout Layers

```python
from tensorflow.keras.layers import Dropout
```

```python
model = Sequential()
model.add(Dense(30, activation='relu'))
model.add(Dropout(0.5))

model.add(Dense(30, activation='relu'))
model.add(Dropout(0.5))
# binary classification
model.add(Dense(1, activation='sigmoid'))
model.compile(loss = 'binary_crossentropy', optimizer = 'adam')
```

```python
model.fit(x= X_train, y=y_train, epochs = 600,validation_data=(X_test, y_test),
         callbacks = [early_stop])
```

```python
model_loss = pd.DataFrame(model.history.history)
```

```python
model_loss.plot()
```

## Model Evaluate

```python
predictions = (model.predict(X_test) > 0.5).astype("int32")
```

```python
predictions
```

```python
from sklearn.metrics import classification_report, confusion_matrix
```

```python
print(classification_report(y_test,predictions ))
```

```python
print(confusion_matrix(y_test, predictions))
```
