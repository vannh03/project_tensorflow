```python
def fill_mort_acc(total_acc,mort_acc):
    
    if np.isnan(mort_acc):
        return total_acc_avg[total_acc]
    else:
        return mort_acc
```

# Creating Model

```python
from sklearn.model_selection import GridSearchCV, RandomizedSearchCV
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Activation,Dropout
from tensorflow.keras.constraints import max_norm
```

### Early Stopping

```python
from tensorflow.keras.callbacks import EarlyStopping
early_stop = EarlyStopping(monitor= 'val_loss', mode = 'min', verbose = 1, patience = 25) # it mean it will stop when accurancy is not decrease
##One of `{"auto", "min", "max"}`. In `min` mode,
 #       training will stop when the quantity
 #       monitored has stopped decreasing; in `"max"`
   #     mode it will stop when the quantity
  #      monitored has stopped increasing; in `"auto"`
 #      mode, the direction is automatically inferred
  #      from the name of the monitored quantity.
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import GridSearchCV
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.regularizers import l2

# Create Model
def create_model(neurons_1, neurons_2, neurons_3, activation, optimizer, dropout):
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Input(shape=78))
    
    model.add(tf.keras.layers.Dense(units=neurons_1,
               activation=activation,kernel_initializer='glorot_uniform'))
    model.add(tf.keras.layers.Dropout(rate=dropout))
    
    model.add(tf.keras.layers.Dense(units=neurons_2,activation=activation))    
    model.add(tf.keras.layers.Dropout(rate=dropout))
    
    model.add(tf.keras.layers.Dense(units=neurons_3,activation=activation))    
    model.add(tf.keras.layers.Dropout(rate=dropout))
    
    model.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
    model.compile(optimizer=optimizer,loss='binary_crossentropy',
                   metrics=['accuracy']) 
    return model
# Create grid
param_grid = {
    'model__activation': ['relu', 'tanh'],
    'model__neurons_1':[78, 85],
    'model__neurons_2':[35,39],
    'model__neurons_3':[19],
    'model__optimizer': ['RMSprop', 'Adam'],
    'batch_size': [256, 512],
    'model__dropout': [0.3],
    'epochs': [25]
}
# set up model
model = KerasClassifier(model=create_model, verbose=2)

# training
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=1, cv=3, verbose=2)
grid_result = grid.fit(X_train, y_train,
                      validation_data=(X_test, y_test),
                      callbacks = [early_stop])
```

```python
print("Best accuracy of: %f using %s" % (grid_result.best_score_, 
                                         grid_result.best_params_))
```

```python
means = grid_result.cv_results_['mean_test_score']
stds = grid_result.cv_results_['std_test_score']
params = grid_result.cv_results_['params']
```

```python
from scikeras.wrappers import KerasClassifier
from sklearn.model_selection import RandomizedSearchCV
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout
from tensorflow.keras.regularizers import l2

def create_model(neurons_1, neurons_2, neurons_3, activation, optimizer, dropout):
    model = tf.keras.Sequential()
    model.add(tf.keras.layers.Input(shape=78))
    
    model.add(tf.keras.layers.Dense(units=neurons_1,
               activation=activation,kernel_initializer='glorot_uniform'))
    model.add(tf.keras.layers.Dropout(rate=dropout))
    
    model.add(tf.keras.layers.Dense(units=neurons_2,activation=activation))    
    model.add(tf.keras.layers.Dropout(rate=dropout))
    
    model.add(tf.keras.layers.Dense(units=neurons_3,activation=activation))    
    model.add(tf.keras.layers.Dropout(rate=dropout))
    
    model.add(tf.keras.layers.Dense(units=1,activation='sigmoid'))
    model.compile(optimizer=optimizer,loss='binary_crossentropy',
                   metrics=['accuracy']) 
    return model

param_grid = {
    'model__activation': ['relu', 'tanh'],
    'model__neurons_1':[78, 85],
    'model__neurons_2':[35,39],
    'model__neurons_3':[19],
    'model__optimizer': ['RMSprop', 'Adam'],
    'batch_size': [256, 512],
    'model__dropout': [0.3],
    'epochs': [25]
}

model = KerasClassifier(model=create_model, verbose=2)

grid = RandomizedSearchCV(estimator=model, param_distributions=param_grid, n_jobs=1, cv=3, verbose=2)
grid_result = grid.fit(X_train, y_train,
                      validation_data=(X_test, y_test),
                      callbacks = [early_stop])
# verbose giúp in ra thông tin trong quá trình training 
# n_jobs : sử dụng cpu
# cv : chia dữ liệu thành các phần, Determines the cross-validation splitting
```
